```python
# Now, plotting each columns of the dataframe value count to know more about the impactful entity in the column
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 30,
        }
l=data.columns.tolist()
for i in range(1,len(l)):
    
    
    if len(data.value_counts(l[i]))>50:
        
        df = data.value_counts(l[i])[:50]
        plt.figure(figsize=(30,10))
        sns.barplot(x=df.index, y=df)
        plt.xticks(rotation=90)
        plt.title(l[i],fontdict=font)
    
    else:
        
        df = data.value_counts(l[i])
        plt.figure(figsize=(30,10))
        sns.barplot(x=df.index, y=df)
        plt.xticks(rotation=90)
        plt.title(l[i],fontdict=font)
```

```python
# picking few columns having high impact 
required_columns = ['VICTIM_COUNT',
    'OFFENSE_NAME', 
    'LOCATION_NAME', 
    'BIAS_DESC', 
    'VICTIM_TYPES',
    'DATA_YEAR', 
    'PUB_AGENCY_NAME',
    'STATE_NAME', 
    'TOTAL_OFFENDER_COUNT','AGENCY_TYPE_NAME', 
    'DIVISION_NAME',
    'REGION_NAME'
    
]
```

```python
data=data[required_columns]
```

```python
sns.heatmap(data.corr(),xticklabels=data.corr().columns.values,yticklabels=data.corr().columns.values,annot=True,cmap="Paired");
```

```python
#Label encoding the columns
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
data['DIVISION_NAME'] = labelencoder.fit_transform(data['DIVISION_NAME'])
data['REGION_NAME'] = labelencoder.fit_transform(data['REGION_NAME'])
data['STATE_NAME'] = labelencoder.fit_transform(data['STATE_NAME'])
data['PUB_AGENCY_NAME'] = labelencoder.fit_transform(data['PUB_AGENCY_NAME'])
data['LOCATION_NAME'] = labelencoder.fit_transform(data['LOCATION_NAME'])
data['BIAS_DESC'] = labelencoder.fit_transform(data['BIAS_DESC'])
data['VICTIM_TYPES'] = labelencoder.fit_transform(data['VICTIM_TYPES'])
data['OFFENSE_NAME'] = labelencoder.fit_transform(data['OFFENSE_NAME'])
data['AGENCY_TYPE_NAME'] = labelencoder.fit_transform(data['AGENCY_TYPE_NAME'])
```

```python
data
```

```python
plt.figure(figsize = (15, 10))
sns.heatmap(data.corr(), cmap="Paired", annot=True, fmt=".1f")
plt.show()
```

```python
#pair ploting bewteen each columns after label encoding

sns.set(rc={'figure.figsize':(100,20)})
sns.pairplot(data);
```
